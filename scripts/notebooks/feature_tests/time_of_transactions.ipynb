```python
# Analysis of correlation between number of transactions and price

from lib.utils import load_all_resale_data, get_cleaned_data
import matplotlib.pyplot as plt
import seaborn as sns

# Load data with transaction date features
X, y = load_all_resale_data(include_features=['month'])
X, y = get_cleaned_data(X, y)

# Create a DataFrame
df = X
df['price'] = y

# show time in relation to price
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df, x='month', y='price', ci=None)
plt.title('Price over Time')
plt.xlabel('Month')
plt.ylabel('Price')
plt.xticks(rotation=45)

df['transactions'] = df.groupby('month')['price'].transform('count')

# show time in relation to number of transactions
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df, x='month', y='transactions', ci=None)
plt.title('Number of Transactions over Time')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# accumulate the number of transactions per month

# create a sns plot with the number of transactions and price
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.scatterplot(data=df, x='transactions', y='price', alpha=0.5)
plt.title('Number of Transactions vs Price')
plt.xlabel('Number of Transactions')
plt.ylabel('Price')
plt.xscale('log')
plt.show()
```
