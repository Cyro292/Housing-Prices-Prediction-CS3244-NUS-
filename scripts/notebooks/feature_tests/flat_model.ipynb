```python
# Let's analyze the 'block' feature and its relationship with housing prices
from lib.utils import load_all_resale_data, get_cleaned_data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data with just the block feature
X, y = load_all_resale_data(include_features=['flat_model'])



# color being the number of values
X['num_values'] = X.groupby('flat_model')['flat_model'].transform('count')

# sort according to mean price
X['mean_price'] = y.groupby(X['flat_model']).transform('mean')
X.sort_values('mean_price', ascending=False, inplace=True)

plt.figure(figsize=(12, 6))
sns.boxplot(x=X['flat_model'], y=y, hue=X['num_values'], palette='viridis')
plt.title('Boxplot of Housing Prices by Flat Model')
plt.xlabel('Flat Model')
plt.ylabel('Housing Price')
plt.xticks(rotation=45)
plt.show()
```
