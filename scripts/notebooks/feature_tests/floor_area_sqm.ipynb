```python
from lib.utils import load_all_resale_data, get_cleaned_data
import seaborn as sns

X, y = load_all_resale_data(include_features=['floor_area_sqm', 'flat_model'])

X, y = get_cleaned_data(X, y)

X['price'] = y

# get sample
X = X.sample(1000)

sns.pointplot(
    x=X["floor_area_sqm"],
    y=X['price'],
)
```

```python
# print the correlation between floor area and price
print("Correlation between floor area and price:")
print(X[['floor_area_sqm', 'price']].corr().loc['floor_area_sqm', 'price'])
```
